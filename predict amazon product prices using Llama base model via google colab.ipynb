## Enhancing Llama Model Efficiency for Predicting Amazon Product Prices Using PEFT, LoRA Adaptors, and QLoRA Quantization Techniques

In this project, we explore advanced techniques to enhance the efficiency of the Llama model for predicting product prices on the Amazon website. By integrating Parameter-Efficient Fine-Tuning (PEFT), Low-Rank Adaptation (LoRA) adaptors, and Quantized Low-Rank Adaptation (QLoRA) quantization techniques, we aim to significantly reduce memory usage while maintaining high performance. Our experiments were conducted on Google Colab, utilizing a hosted runtime with an NVIDIA T4 GPU. The results demonstrate that quantization techniques can effectively lower memory requirements, making the model more efficient and scalable for practical applications. This project highlights the potential of combining PEFT, LoRA, and QLoRA to optimize machine learning models for real-world tasks.

```python
# Each of the Target Modules has 2 LoRA Adaptor matrices, called lora_A and lora_B
# These are designed so that weights can be adapted by adding alpha * lora_A * lora_B
# Let's count the number of weights using their dimensions:

# See the matrix dimensions above
lora_q_proj = 4096 * 32 + 4096 * 32
lora_k_proj = 4096 * 32 + 1024 * 32
lora_v_proj = 4096 * 32 + 1024 * 32
lora_o_proj = 4096 * 32 + 4096 * 32

# Each layer comes to
lora_layer = lora_q_proj + lora_k_proj + lora_v_proj + lora_o_proj

# There are 32 layers
params = lora_layer * 32

# So the total size in MB is
size = (params * 4) / 1_000_000

print(f"Total number of params: {params:,} and size {size:,.1f}MB")
```
